Next we consider a dataset of face images,each  32x32 in greyscale. The first hundred faces are dislayed below.

```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
from PIL import Image
import scipy.io
```

```python
%matplotlib inline
file='ex7faces.mat'
tmp=scipy.io.loadmat(file)
X=tmp['X']
fig, axs = plt.subplots(10, 10)
for i in range(10):
    for j in range(10):
        tmp2=(X[10*i +j ,:].reshape(32,32)).T
        axs[i,j].imshow(tmp2, cmap='gray_r') 
        axs[i,j].axis('off') 
```
